```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
        'Increase': [0.5, 0.4, 0.7, 1.2, 0.8, 1.5, 0.9, 1.3, 1.1, 0.3, 0.6, 3.2]}

transport_df = pd.DataFrame(data)

```

```python
transport_df
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.figure(figsize=(10,6))
plt.plot(transport_df['Month'], transport_df['Increase'], marker='o', color='b')

plt.title('Monthly Transportation Price Increases in Canada (2023)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Increase', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python


plt.figure(figsize=(8,8))
plt.pie(transport_df['Increase'], labels=transport_df['Month'], autopct='%1.1f%%', startangle=140)

plt.title('Transportation Price Increases in Canada (2023)', fontsize=14)

plt.show()
```
